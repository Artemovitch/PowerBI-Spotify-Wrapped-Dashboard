```python
import pandas as pd
import json
import glob
```

```python
json_files = glob.glob('my_spotify_data/Streaming_History_Audio_*.json')
```

```python
print(len(json_files))
```

```python
dfs = []
for file in json_files:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        dfs.append(pd.DataFrame(data))
```

```python
full_history = pd.concat(dfs, ignore_index=True)
```

```python
full_history.to_csv('my_spotify_data.csv', index=False)
print(f"Combined {len(json_files)} files into one DataFrame with {len(full_history)} rows.")
```
